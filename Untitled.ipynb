```python
from craigslist import CraigslistForSale
```

```python
CraigslistForSale.show_filters(category="cta")
```

```python
cl_c = CraigslistForSale(site='baltimore',
                  category='cta',
                  filters={'search_distance': 25,
                           'zip_code': 21210,
                           'min_price': 3500,
                           'max_price': 8000,
                           'max_miles': 100000,
                           'min_year': 2009
                          })
```

```python
results = cl_c.get_results(sort_by='price_asc')
```

```python
import pprint
pp = pprint.PrettyPrinter(indent=4)
```

```python
list_dict = []
for result in results:
    pp.pprint(result)
    list_dict.append(result)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list_dict)
```

```python
cols = df.columns.tolist()
```

```python
cols = ['name',
 'price',
'datetime',
 'last_updated',
 'where',
 'url',
 'geotag',
 'has_image',
 'id',
 'repost_of'
 ]
```

```python
df = df[cols]
```
